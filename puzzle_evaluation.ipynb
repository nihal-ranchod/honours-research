```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import json
```

Extracting Relevant Puzzle Data

```python
with open('puzzle_evaluation_results.json', 'r') as file:
    data = json.load(file)

bot_results = data['puzzle_evaluation_summary']
bots = list(bot_results.keys())
solve_rates = [bot_results[bot]['solve_rate'] for bot in bots]
move_accuracies = [bot_results[bot]['move_accuracy'] for bot in bots]
avg_times_per_puzzle = [bot_results[bot]['avg_time_per_puzzle'] for bot in bots]
```

Creating DataFrame

```python
df = pd.DataFrame({
    'Bot': bots,
    'Solve Rate': solve_rates,
    'Move Accuracy': move_accuracies,
    'Avg Time per Puzzle': avg_times_per_puzzle
})
display(df)
```

Plotting Solve Rates

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Bot'], df['Solve Rate'], color='skyblue')
plt.title('Solve Rates of Bots')
plt.ylabel('Solve Rate')
plt.xlabel('Bot')
plt.xticks(rotation=45)
plt.savefig("Puzzle Evaluation Plots/solve_rates.png", bbox_inches='tight')
plt.show()
```

Plotting Move Accuracies

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Bot'], df['Move Accuracy'], color='lightgreen')
plt.title('Move Accuracies of Bots')
plt.ylabel('Move Accuracy')
plt.xlabel('Bot')
plt.xticks(rotation=45)
plt.savefig("Puzzle Evaluation Plots/move_accuracies.png", bbox_inches='tight')
plt.show()
```

Plotting Avg Time per Puzzle

```python
plt.figure(figsize=(10, 5))
plt.bar(df['Bot'], df['Avg Time per Puzzle'], color='salmon')
plt.title('Average Time per Puzzle of Bots')
plt.ylabel('Avg Time per Puzzle (seconds)')
plt.xlabel('Bot')
plt.xticks(rotation=45)
plt.savefig("Puzzle Evaluation Plots/avg_time_per_puzzle.png", bbox_inches='tight')
plt.show()
```
